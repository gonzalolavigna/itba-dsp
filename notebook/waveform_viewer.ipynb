```python
import matplotlib.pyplot as plt
import numpy as np

width = 40
height = 20

import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [width, height]

data     = np.load('../output/signal_input.npy')
data_fxp = np.load('../output/signal_fxp.npy')

print("{:2.2f}".format(data_fxp[0]))
print("{:2.2f}".format(data_fxp[25]))
print("{:2.2f}".format(data_fxp[50]))
print("{:2.2f}".format(data_fxp[75]))
print("{:2.2f}".format(data_fxp[99]))
print(data.size)
```

```python
fig, ax = plt.subplots(1, 1)
ax.plot(data)
ax.plot(data_fxp)
ax.set_xlabel('Samples [N]')
ax.set_ylabel('Magnitud')
ax.set_title('Señal numpy')
ax.grid()
```

```python
fig, ax = plt.subplots(1, 1)
ax.plot(np.abs(data-data_fxp))
ax.set_xlabel('Samples [N]')
ax.set_ylabel('Magnitud')
ax.set_title('Señal numpy')
ax.grid()

print(np.max(np.abs(data-data_fxp)))
```
